```python
import pandas as pd
import csv

base_file = '../data/weather/air_temp/air_temp'
target_file = '../data/weather/air_temp/avg_air_temp'
tree_file = '../models/abiesAlba.csv'
target_rain = '../models/output/avg_rain_tree.csv'

def gen_temp_df(file_name, year):
    df = pd.read_csv(file_name, delimiter= r"\s+")
    avg_df = df.iloc[:, [0,1]].copy()
    avg_df[year + '_avg'] = df.mean(axis = 1) 
    avg_df[year + '_avg'] = avg_df[year + '_avg'].abs()
    return avg_df
def gen_base_df(file_name):
    df = pd.read_csv(file_name, delimiter= r"\s+")
    avg_df = df.iloc[:, [0,1]].copy()
    return avg_df

def get_all_avg(base_file, num_years, begin_year):
    sum_df = gen_base_df(base_file + str(begin_year))
    for i in range(begin_year, begin_year + num_years):
        temp_df = gen_temp_df(base_file + str(i), str(i))
        sum_df = pd.merge(sum_df, temp_df, on = ['Longitude', 'Latitude'], how = 'left')
    return sum_df

def avgAllYrs(df):
    ans = df.iloc[:, [0,1]].copy() 
    ans['avg_rainfall'] = df.mean(axis = 1) 
    return ans
```

```python
file_name = '../data/weather/air_temp/air_temp2000'
gen_temp_df(file_name, '2020')
```

```python
all_avg_df = get_all_avg(base_file, 10, 2000)
all_avg_df.head()
```

```python
ans_df = avgAllYrs(all_avg_df)
ans_df.to_csv(target_file, sep='\t')
```

```python
def combine_models(avg_tempdf, tree_df):
    return pd.merge(avg_tempdf, tree_df, on = ['Longitude', 'Latitude'], how = 'right')
tree_df = pd.read_csv(tree_file, delimiter= ',')
tree_df['Latitude'] = tree_df['Latitude'].apply(lambda x: round(x, 0) + .25)
tree_df['Longitude'] = tree_df['Longitude'].apply(lambda x: round(x, 0) +.25)
updated_df = combine_models(ans_df, tree_df)
updated_df.to_csv(target_rain, sep=',')
updated_df.head()
```
